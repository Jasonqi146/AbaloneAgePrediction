```python
import csv
f = open("drive/My Drive/CX4240 Project/Abalone.csv",)
reader = csv.reader(f)
data = list(reader)

print(len(data))
data_body = data[1:]
f.close()
```

```python
gender = []
count = [0,0,0]
for i in range(len(data_body)):
  if data_body[i][0] == 'F':
    gender.append([1, 0, 0])
    count[0] += 1
  elif data_body[i][0] == 'M':
    gender.append([0, 1, 0])
    count[1] += 1
  else:
    gender.append([0, 0, 1])
    count[2] += 1
  #print(data_body[i][0])
print(count)
```

```python
other = []
for i in range(len(data_body)):
  lis = [float(data_body[i][j]) for j in range(1,len(data_body[i]))]
  other.append(lis)
for i in other:
  print(i)
```

```python
import numpy as np
print(np.array(gender).shape)
print(np.array(other).shape)
good_data = np.concatenate((np.array(gender), np.array(other)), axis=1)
print(good_data.shape)
```

```python
train = good_data[0:3133]
test = good_data[3133:]
print(len(train))
print(len(test))
```

```python
x_train = train[:,0:10]
y_train = train[:,10]
x_test = test[:,0:10]
y_test = test[:,10]
```

```python
print(y_train)
```

```python
for i in range(len(y_train)):
  if y_train[i] < 9:
    y_train[i] = 0
  elif y_train[i] < 11:
    y_train[i] = 1
  else:
    y_train[i] = 2
for i in range(len(y_test)):
  if y_test[i] < 9:
    y_test[i] = 0
  elif y_test[i] < 11:
    y_test[i] = 1
  else:
    y_test[i] = 2
```

Random Forest Fine-tuning

```python
from sklearn.ensemble import RandomForestClassifier
#from sklearn.ensemble import RandomForestRegressor
rf = RandomForestClassifier(random_state = 42)
#rf = RandomForestRegressor(random_state = 42)
from pprint import pprint
# Look at parameters used by our current forest
print('Parameters currently in use:\n')
pprint(rf.get_params())
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 600, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt', 'log2']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 20]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
#oob_score
oob_score = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap,
               'oob_score': oob_score}
pprint(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestClassifier()
#rf = RandomForestRegressor()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 200, cv = 5, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(x_train, y_train)
```

```python
rf_random.best_params_
```

```python
def evaluate(model, test_features, test_labels):
    score = model.score(test_features, test_labels)
    print("score:",  score)
    return score
```

```python
base_model = RandomForestRegressor(n_estimators = 100, random_state = 42)
base_model.fit(x_train, y_train)
base_R2 = evaluate(base_model, x_test, y_test)
best_random = rf_random.best_estimator_
random_R2 = evaluate(best_random, x_test, y_test)
```

```python
print('Improvement of {:0.2f}%.'.format( 100 * (random_R2 - base_R2) / base_R2))
```
